# Sorting with NumPy



Intel Publishes Blazing Fast [Intel AVX-512](https://www.phoronix.com/news/Intel-AVX-512-Quicksort-Numpy) **Sorting Library, Numpy** Switching To It For 10~17x Faster Sorts

Here we will compare quicksort, mergesort and heapsort both strqaighforward loop implementations versus NumPy equivalents


````python
import numpy as np
import time
BIG = 1_000_000
np.random.seed(seed=12)
arr = np.random.rand(BIG)
orig = arr.copy()
````

````python
def heapify(arr, n, i):
    largest = i 
    l = 2 * i + 1 
    r = 2 * i + 2 
 
    if l < n and arr[largest] < arr[l]:
        largest = l
 
    if r < n and arr[largest] < arr[r]:
        largest = r
 
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  
        heapify(arr, n, largest)
 
def heapSort(arr):
    n = len(arr)
 
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
 
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i] 
        heapify(arr, i, 0)
    return arr
````

````python
np.random.seed(seed=12)
arr = np.random.rand(BIG)
timing = {}
t1 = time.time()
heapSort(arr)
t2 = time.time()
print("Sorted array is:")
print(arr[:10] )
timing['heapsort_bruteForce'] = time.time() - t1
print('Elapsed time', timing['heapsort_bruteForce'])
````

````python
def quickSort(arr, low, high):
    if low < high:
        pivotIndex = partition(arr, low, high)
        quickSort(arr, low, pivotIndex - 1)
        quickSort(arr, pivotIndex + 1, high)

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1  # Index of smaller element
    for j in range(low, high):
        # If current element is smaller than or equal to pivot
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i + 1
````

# Mergesort


````python
def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        leftHalf = arr[:mid]
        rightHalf = arr[mid:]
        
        mergeSort(leftHalf)
        mergeSort(rightHalf)
        
        i = j = k = 0
        
        while i < len(leftHalf) and j < len(rightHalf):
            if leftHalf[i] < rightHalf[j]:
                arr[k] = leftHalf[i]
                i += 1
            else:
                arr[k] = rightHalf[j]
                j += 1
            k += 1
        
        while i < len(leftHalf):
            arr[k] = leftHalf[i]
            i += 1
            k += 1
        
        while j < len(rightHalf):
            arr[k] = rightHalf[j]
            j += 1
            k += 1
````

````python
import time
np.random.seed(seed=12)
arr = np.random.rand(BIG)
timing = {}
t1 = time.time()
mergeSort(arr)
t2 = time.time()
print("Sorted array is:")
timing['mergesort_bruteForce'] = time.time() - t1
print('Elapsed time', timing['mergesort_bruteForce'])
````
